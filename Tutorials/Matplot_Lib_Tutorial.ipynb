```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = (1, 2, 3, 4, 5)
y = (1, 2, 2.5, 4, 5)

plt.plot(x, y)
plt.show()
```

```python
xpoints = [1, 2, 3, 5, 8]
ypoints = [5, 3, 2, 1, 1]

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
x = ['apple', 'banana', 'orange', 'strawberry']
calories = [95, 105, 62, 49]

plt.bar(x, calories)

plt.show()
```

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```
