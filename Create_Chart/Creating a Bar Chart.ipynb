# Creating a Bar Chart
A **bar chart**, also known as a bar graph, is a common type of data visualization used to represent categorical data. It displays data using rectangular bars, where the length or height of each bar is proportional to the quantity or value it represents. Bar charts are particularly effective for comparing the values of different categories or groups.

In this tutorial, we will create a bar chart to visualize and compare the GDP of the top 10 countries in 2020.

### Step 1: Import the Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2: Load and Explore the Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/yangliuiuk/data/main/gdp.csv')
df.head()
```

### Step 3: Extract Data for the Year 2020
Extract the GDP data for the year 2020 and sort the DataFrame in descending order based on GDP values.

```python
# Extract GDP data for the year 2020 and sort by GDP in descending order
df_2020 = df[['Country Name', '2020']].sort_values(by='2020', ascending=False)
df_2020[0:50]
#df_2020.iloc[0:50, :]
```

Here note that the dataset includes regions as country names. So, we need to examine the first number of rows in order to select actual countries. 

The top 10 countries with highest GDP are:

- United States
- China
- Japan
- Germany
- United Kingdom
- India
- France
- Italy
- Canada
- Korea, Rep.

### Step 4: Select the Top 10 GDP Countries

```python
# List of the 10 countries you want to compare
countries_to_compare = [
    'United States',
    'China',
    'Japan',
    'Germany',
    'United Kingdom',
    'India',
    'France',
    'Italy',
    'Canada',
    'Korea, Rep.'
]

# Filter the DataFrame to include only the selected countries
selected_countries_df = df[df['Country Name'].isin(countries_to_compare)] #Selects the rows of the countries
#df[logical expression] selects rows with true values. Logical expression gives array of true false
selected_countries_df = selected_countries_df[['Country Name', '2020']].sort_values(by='2020', ascending=False)
selected_countries_df
```

### Step 5: Create the Bar Chart

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Convert GDP values to billions
selected_countries_df['GDP (Billions)'] = selected_countries_df['2020'] / 1e9  # Divide by 1 billion
selected_countries_df
```

```python
# Create the bar chart
plt.bar(selected_countries_df['Country Name'], selected_countries_df['GDP (Billions)'], color='blue', alpha=0.7)
plt.xlabel('Country')
plt.ylabel('GDP (Billions)')
plt.title('GDP Comparison of Selected Countries in 2020')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Show the bar chart
plt.tight_layout()
plt.show()
```
