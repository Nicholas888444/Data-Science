# Creating a Line Plot
In this tutorial, we will create a line plot on GDP data.  
A line plot, also known as a line chart or line graph, is a type of data visualization used primarily for showing the relationship and trends between two continuous variables. 

### Step 0: Import Required Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 1: Load the CSV data into a Pandas DataFrame
In this example, we will use the GDP dataset from the link below. The GDP dataset includes information about the Gross Domestic Product (GDP) of different countries from 1960 to 2020.   
The dataset can be used to analyze and track the economic performance of different countries or regions over time. It can be used for various economic and statistical analyses, including trend analysis, comparisons between countries, and more.

```python
df = pd.read_csv('https://raw.githubusercontent.com/yangliuiuk/data/main/gdp.csv')
```

Explore the dataset by df.head().

```python
df.head()
```

Columns:
- 'Country Name': The name of the country.
- 'Code': A country code or identifier.
- '1960 to 2020': These columns represent GDP values for each country in the corresponding year. 

### Step 2: Filter and extract United States GDP data
We will first create a plot on US GDP data.

```python
us_gdp = df[df['Country Name'] == 'United States']
```

```python
us_gdp_values = us_gdp.loc[:, '1960':'2020'].iloc[0, :]
```

```python
us_gdp_values_in_billions = us_gdp_values / 1e9  # Divide by 1 billion
```

### Step 3: Create a Matplotlib line plot to visualize the data

```python
years = range(1960, 2021)
```

```python
plt.figure(figsize=(12, 6))  # Set the figure size
plt.plot(years, us_gdp_values_in_billions, marker='o', linestyle='-', color='b', label='United States GDP (Billions)')

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('GDP Value (Billions)')
plt.title('United States GDP from 1960 to 2020 (Billions)')

# Add gridlines
plt.grid(True)

# Display the plot
plt.legend()
plt.tight_layout()  # Adjust layout
plt.show()
```

### Step 4: Compare US and Canada GDP
We can plot to series on a single line chart to compare the two series. For example, we can compare US GDP with Canada GDP.

```python
# Getting US GDP
us_gdp = df[df['Country Name'] == 'United States']
us_gdp_values = us_gdp.loc[:, '1960':'2020'].iloc[0, :]
us_gdp_values_in_billions = us_gdp_values / 1e9  # Divide by 1 billion

# Getting Canada GDP
canada_gdp = df[df['Country Name'] == 'Canada']
canada_gdp_values = canada_gdp.loc[:, '1960':'2020'].iloc[0, :]
canada_gdp_values_in_billions = canada_gdp_values / 1e9  # Divide by 1 billion

years = range(1960, 2021)
plt.figure(figsize=(12, 6))  # Set the figure size
plt.plot(years, us_gdp_values, marker='o', linestyle='-', color='b', label='United States GDP')
plt.plot(years, canada_gdp_values, marker='s', linestyle='-', color='r', label='Canada GDP')

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('GDP Value')
plt.title('Comparison of United States and Canada GDP from 1960 to 2020')

# Add gridlines
plt.grid(True)

# Display the legend
plt.legend()

# Display the plot
plt.tight_layout()  # Adjust layout
plt.show()
```

### Step 5: Drawing Conclusions:
Plotting and comparing the GDP series of the United States and Canada in a line chart allows us to visually analyze and draw several conclusions from the data:
 - **Relative Economic Size**: The line chart will clearly show that the GDP of the United States is significantly larger than that of Canada throughout the entire period from 1960 to 2020. The gap between the two economies is quite substantial.
 - **Economic Growth Trends**: By examining the slopes of the lines, we can observe the growth trends of both countries. For example, it will be evident that the United States has consistently experienced higher GDP growth rates compared to Canada, resulting in a widening gap in economic size over time.
 - **Economic Events**: Sudden changes or fluctuations in the lines may correspond to significant economic events or crises. For instance, the financial crisis of 2008-2009 is likely to be noticeable in both GDP series.
 - **Impact of COVID-19 Pandemic**: The chart can show the impact of the COVID-19 pandemic on both economies, with a likely sharp decline in GDP in 2020.
