# Creating a Pie Chart
A pie chart is a type of graphical representation used to display data as a circular graph, where the whole circle represents 100% or the total, and the individual segments, or "slices," represent parts or percentages of the whole. Pie charts are commonly used to visualize data that can be divided into categories or components. Each slice of the pie represents the proportion or percentage of a specific category relative to the total.

In this tutorial, we will explore the Titanic Disaster Dataset. 
The Titanic Disaster Dataset is a well-known dataset in the field of data science and machine learning. It contains information about the passengers who were aboard the RMS Titanic during its ill-fated maiden voyage in April 1912. The dataset is often used for data analysis, visualization, and predictive modeling tasks, making it a popular choice for educational purposes and data science competitions.

The dataset typically includes the following information for each passenger:

Survived: A binary variable indicating whether the passenger survived (1) or did not survive (0).

Pclass: The passenger's class (1st, 2nd, or 3rd class).

Name: The passenger's name.

Sex: The gender of the passenger (male or female).

Age: The age of the passenger.

Siblings/Spouses Aboard: The number of siblings or spouses the passenger had aboard.

Parents/Children Aboard: The number of parents or children the passenger had aboard.

Fare: The fare paid by the passenger.

The dataset can be found at https://www.kaggle.com/c/titanic/data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/yangliuiuk/data/main/titanic.csv")
df.head()
```

### Create a pie chart that shows the distribution of survival status (Survived vs. Did not survive)

```python
# Calculate the count of survivors and non-survivors
survived_count = df['Survived'].value_counts()
survived_count
```

```python
# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(survived_count, labels=['Did not survive', 'Survived'], autopct='%1.1f%%', startangle=90, shadow=False, colors=['lightcoral', 'lightgreen'])
# We use autopct to display the percentage on each slice, startangle to set the starting angle of the pie chart, and shadow=True for a shadow effect.
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Survival Status Distribution', fontsize=16, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the pie chart
plt.show()
```

### Create a pie chart that shows the distribution of passenger class.

```python
pclass_count = df['Pclass'].value_counts()
pclass_count
```

```python
# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(pclass_count, labels=['Class 3', 'Class 1', 'Class 2'], autopct='%1.1f%%', startangle=90, shadow=True, colors=['lightcoral', 'lightgreen', 'lightblue'])
# We use autopct to display the percentage on each slice, startangle to set the starting angle of the pie chart, and shadow=True for a shadow effect.
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Passenger Class Distribution', fontsize=16, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the pie chart
plt.show()
```

### Create a pie chart that shows the distribution of gender.

```python
sex_count = df['Sex'].value_counts()
sex_count
```

```python
# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(sex_count, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90, shadow=True, colors=['lightcoral', 'lightgreen'])
# We use autopct to display the percentage on each slice, startangle to set the starting angle of the pie chart, and shadow=True for a shadow effect.
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Gender Distribution', fontsize=16, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the pie chart
plt.show()
```

### Create two pie charts side by side to show the gender distribution of the "Survived" and "Did not survive" groups.

```python
# Create subplots for two pie charts side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Filter data for survived and not survived groups
survived_data = df[df['Survived'] == 1]
not_survived_data = df[df['Survived'] == 0]

# Calculate gender counts for both groups
survived_gender_count = survived_data['Sex'].value_counts()
not_survived_gender_count = not_survived_data['Sex'].value_counts()

# Pie chart 1: Gender distribution of Survived group
axs[0].pie(survived_gender_count, labels=survived_gender_count.index, autopct='%1.1f%%', startangle=90, shadow=True, colors=['lightblue', 'lightpink'])
axs[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[0].set_title('Gender Distribution (Survived)', fontsize=14, fontweight='bold')

# Pie chart 2: Gender distribution of Not Survived group
axs[1].pie(not_survived_gender_count, labels=not_survived_gender_count.index, autopct='%1.1f%%', startangle=90, shadow=True, colors=['lightcoral', 'lightgreen'])
axs[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[1].set_title('Gender Distribution (Did not survive)', fontsize=14, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the pie charts
plt.show()
```

### Create two pie charts side by side to show the passenger class distribution of the "Survived" and "Did not survive" groups.

```python
# Create subplots for two pie charts side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Filter data for survived and not survived groups
survived_data = df[df['Survived'] == 1]
not_survived_data = df[df['Survived'] == 0]

# Calculate passenger class counts for both groups
survived_pclass_count = survived_data['Pclass'].value_counts()
not_survived_pclass_count = not_survived_data['Pclass'].value_counts()

# Define labels for passenger classes
pclass_labels = ['Class 3', 'Class 1', 'Class 2']

# Pie chart 1: Passenger class distribution of Survived group
axs[0].pie(survived_pclass_count, labels=pclass_labels, autopct='%1.1f%%', startangle=90, shadow=True, colors=['lightblue', 'lightpink', 'lightgreen'])
axs[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[0].set_title('Passenger Class Distribution (Survived)', fontsize=14, fontweight='bold')

# Pie chart 2: Passenger class distribution of Not Survived group
axs[1].pie(not_survived_pclass_count, labels=pclass_labels, autopct='%1.1f%%', startangle=90, shadow=True, colors=['lightcoral', 'lightyellow', 'lightgray'])
axs[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[1].set_title('Passenger Class Distribution (Did not survive)', fontsize=14, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the pie charts
plt.show()
```
