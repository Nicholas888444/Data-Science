# Creating a Histogram
A histogram is a graphical representation of the distribution of a dataset. It is used to visualize the frequency or count of data points falling into specific intervals or "bins" within a continuous range of values. Histograms are particularly useful for understanding the shape, central tendency, and spread of data.

In this tutorial, we will learn how to create a histogram showing the GDP distribution for the year 2020.

### Step 1: Import the Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2: Load and Expore the GDP Data

```python
df = pd.read_csv('https://raw.githubusercontent.com/yangliuiuk/data/main/gdp.csv')
# Extract GDP data for the year 2020
gdp_2020 = df['2020'] / 1e9
gdp_2020.head()
```

### Step 3: Create the Histogram

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the histogram
plt.hist(gdp_2020, bins=10, color='blue', alpha=0.7)
plt.xlabel('GDP (2020)')
plt.ylabel('Number of Countries/Regions')
plt.title('GDP Distribution in 2020')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the histogram
plt.tight_layout()
plt.show()
```

### Step 4: Interpret the Results

The histrogram shows a skewed right distribution, which suggests that most countries have GDP values lower than the median, with a long tail on the right.

A wide distribution with greater variability suggests that GDP values among the countries vary significantly, indicating a diverse economic landscape.

A high frequency in a particular bin  indicates that more countries fall within that GDP range.
