```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Download the Dataset
The dataset was taken from https://www.kaggle.com/datasets/aakashjoshi123/virtual-reality-experiences

The dataset goes over motion sickness and immersion levels of people using VR headsets. Motion sickness is measured from 10 (Very motion sick) to 1 (Not motion sick). Immersion level is measured from 5 to 1 with 5 being fully immersed.

```python
#Load the dataframe
df = pd.read_csv("data.csv")
df.head()
```

Using this dataset, I will first see which headset is the most common

```python
df['VRHeadset'].value_counts()
```

```python
# Create the bar chart
plt.bar(['Oculus Rift', 'HTC Vive', 'Playstation VR'], df['VRHeadset'].value_counts(), color=['blue', 'green', 'pink'], alpha=0.7)
plt.xlabel('VR Headset')
plt.ylabel('Headset Count')
plt.title('Headsets included in the study')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Show the bar chart
plt.tight_layout()
plt.show()
```

This bar chart shows that the Oculus Rift is the most popular headset, followed by the HTC Vive, then followed by the Playstation VR. However, there is no large difference in the headsets used in this study. Most headsets appear to be mostly equal.

```python
#Seperate the dataframe by headset
df_oculus = df[df['VRHeadset'] == 'Oculus Rift']
df_vive = df[df['VRHeadset'] == 'HTC Vive']
df_ps = df[df['VRHeadset'] == 'PlayStation VR']

df_oculus['ImmersionLevel'].value_counts().sort_index()
```

```python
# Create subplots for two pie charts side by side
fig, axs = plt.subplots(1, 3, figsize=(12, 6))

# Set the figure size
plt.figure(figsize=(10, 6))

# Pie chart 1: Oculus Rift Immersion Levels
axs[0].pie(df_oculus['ImmersionLevel'].value_counts().sort_index(), labels=[1, 2, 3, 4, 5], autopct='%1.1f%%', startangle=90, shadow=False, colors=['lightcoral', 'lightsalmon', 'gold', 'lawngreen', 'royalblue'])
axs[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[0].set_title('Immersion(Oculus)', fontsize=14, fontweight='bold')

# Pie chart 2: HTC Vive Immersion Levels
axs[1].pie(df_vive['ImmersionLevel'].value_counts().sort_index(), labels=[1, 2, 3, 4, 5], autopct='%1.1f%%', startangle=90, shadow=False, colors=['lightcoral', 'lightsalmon', 'gold', 'lawngreen', 'royalblue'])
axs[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[1].set_title('Immersion(HTC Vive)', fontsize=14, fontweight='bold')

# Pie chart 3: Playstation VR Immersion Levels
axs[2].pie(df_ps['ImmersionLevel'].value_counts().sort_index(), labels=[1, 2, 3, 4, 5], autopct='%1.1f%%', startangle=90, shadow=False, colors=['lightcoral', 'lightsalmon', 'gold', 'lawngreen', 'royalblue'])
axs[2].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[2].set_title('Immersion(Playstation VR)', fontsize=14, fontweight='bold')

# Show the histogram
plt.tight_layout()
plt.show()
```

Looking at each pie chart, the levels of immersion are similar. The highest levels of immersion actually go to Playstation VR with 19.8% of users experiencing maximum immersion and 21.1% of the users experienced very immeresed. HTC Vive has the higher number of less immersed people in categories 1 and 2. Oculus users expereinced the highest levels of middle level immersion of 3 and 4.

The 3 headsets appear to be relatively equal in terms of immersion, with no one headset being significantly better than the other in terms of immersion

```python
# Create subplots for two pie charts side by side
fig, axs = plt.subplots(1, 3, figsize=(12, 6))

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the histogram
axs[0].hist(df_oculus['MotionSickness'], bins=10, color='blue', alpha=0.7)
axs[0].set_title('Oculus', fontsize=14, fontweight='bold')

axs[1].hist(df_vive['MotionSickness'], bins=10, color='green', alpha=0.7)
axs[1].set_title('HTC Vive', fontsize=14, fontweight='bold')

axs[2].hist(df_ps['MotionSickness'], bins=10, color='pink', alpha=0.7)
axs[2].set_title('Playstation VR', fontsize=14, fontweight='bold')

# Show the histogram
plt.tight_layout()
plt.show()
```

These histograms provide interesting information about each haeadset. The Playstation VR headset seems to generally follow a normal distribution pattern with more people feeling average levels of motion sickness. The HTC Vive headset follows a trend of most bins being equal with the number of people in them. The Oculus has a large portion of people in the 7 level of motion sickness, but the general trend of the chart is less people are motion sick. 

The data provided is rather even between headsets. It is hard to draw any meaningful conclusions from this dataset. One conclusion that I can draw is that these three headsets provide no additional levels of immersion or levels of motion sickness compared to each other based on this data.
